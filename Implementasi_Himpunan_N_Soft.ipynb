```python
import pandas as pd
```

Data

```python
data1 = pd.read_excel('/content/drive/MyDrive/Colab Data/Data/Data1.xlsx')
bobot = pd.read_excel('/content/drive/MyDrive/Colab Data/Data/bobot.xlsx')
```

Fungsi untuk menghitung EWCV

```python
def func1(a,b):
  baris = []
  for i in range(len(a)):
      baris.append([a['U'][i],a['a1'][i]*bobot[b][0],a['a2'][i]*bobot[b][1],a['a3'][i]*bobot[b][2],a['a4'][i]*bobot[b][3],a['a5'][i]*bobot[b][4]])
  dataframe = pd.DataFrame(baris, columns=['U','a1', 'a2','a3','a4','a5'])
  sumlist = ['a1','a2','a3','a4','a5']
  dataframe['ewcv'] = dataframe[sumlist].sum(axis=1)
  return dataframe
```

Fungsi untuk mengurutkan data

```python
def func2(data):
  return data.sort_values(by="ewcv", ascending=False).drop(columns=['a1', 'a2','a3','a4','a5'])
```
